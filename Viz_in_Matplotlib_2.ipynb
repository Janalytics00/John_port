```python
# Required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the data
x = [1,2,3,4,5]
y = [100,200,150,300,250]

# creating a basic line chart

plt.plot(x,y)
plt.title('Basic Line Chart')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python

# Define the data
x = [1,2,3,4,5]
y = [100,200,150,300,250]

# creating a basic line chart
# With 'o' market
plt.plot(x,y, marker = 'o')
plt.title('Basic Line Chart')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
# With '*' market
plt.plot(x,y, marker = '*')
plt.title('Basic Line Chart')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
# With '>' market
plt.plot(x,y, marker = '>')
plt.title('Line Chart with Marker')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
# Create a line chart with different line styles
plt.plot(x,y, '-o', color = 'red', linewidth = 2, label = 'line-1')
plt.plot(x, [200,230,250,200,150], '--', color = 'blue', linewidth = 4, label = 'line-2')
plt.title('Line Chart with Different line style')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# creating a table
data = {
    'Months':['Jan', 'Feb', 'Mar', 'Apr', 'May','Jun', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec'],
    'Sales':[100, 120, 130, 140, 150, 160, 145, 135, 155, 165, 170, 190],
    'Profit':[80, 60, 100, 70, 100, 120, 80, 90, 100, 150, 155, 160 ]
}
# Creating a DataFrame
sales = pd.DataFrame(data)
```

```python
sales.head()
```

```python
fig, ax = plt.subplots(figsize = (12,4))
ax.bar(sales['Months'], sales['Sales'], color = 'darkblue', label = "Sales")
ax.plot(sales['Months'], sales['Profit'], color = 'red', label = 'Profit')

ax.set_xlabel('Month')
ax.set_ylabel('Amount(in USD)')
ax.legend(loc = 'upper left')
plt.show()
```

```python
sales = [100,200,300,400,500]
quantity = [10,20,30,40,50]
profit = [90,160,210,280,400]
plt.style.use('default')

fig, ax = plt.subplots(figsize=(10, 4))
ax.bar(np.arange(len(sales)), sales, align = 'center')

# plot the line chart for quantity
ax.plot(np.arange(len(quantity)), quantity, color = 'red', linestyle = 'dashed')

# plot the line chart for profit
ax.plot(np.arange(len(profit)), profit, color = 'orange', linestyle = 'dotted')

# adding labels and annotations
ax.set_xlabel('Month')
ax.set_ylabel('Value')
ax.set_xticks(np.arange(len(sales)))
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May'])
ax.legend(['Quantity', 'Profit'])

# adding a data label
for i, v in enumerate(sales):
  ax.text(i, v, str(v), ha = 'center', va = 'bottom')

plt.show()
```

```python
target = 1000
sales = [920, 900, 1100, 1050, 890, 1090, 950, 1020]
plt.plot([target] * len(sales), label = 'Target')
plt.plot(sales, label = 'Sales')
plt.legend()
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Target vs Sales')
plt.show()
```

```python
lc3 = pd.read_excel("/content/Data (1).xlsx", sheet_name="LC3")
```

```python
print(plt.style.available)
```

```python
month = lc3.Month
unitSold = lc3.UnitsSold
total_transaction = lc3.TotalTransaction
 # create figure and axis object
fig, ax = plt.subplots(figsize = (20,7))


plt.style.use('seaborn-dark')
ax.bar(month, total_transaction, align='center', color = 'darkblue', edgecolor = 'white',
      label = 'Total Transaction')

ax.set_ylabel('Total Transaction', color = 'darkblue')
ax.tick_params(axis = 'y', labelcolor = 'darkblue')

# Change Axis :

ax1 = ax.twinx()

ax1.plot(month, unitSold, color = 'red', linewidth = 4, label = "Units Sold")
ax1.set_ylabel('Units Sold', color = 'red')
ax1.tick_params(axis = 'y', labelcolor = 'red')

for i , v in enumerate(unitSold):
    ax1.text(i, v, str(v), color = 'red', ha = 'center', va = "bottom", fontsize = 'x-large')

ax.legend(loc = 'upper left')
ax1.legend(loc = 'upper right')
plt.show()
```
