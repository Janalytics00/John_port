### Find Out Number of courses per subjects

```python
import matplotlib.pyplot as plt
```

```python
data = data.copy()
data['subject'] = data['subject'].astype('category')
```

```python
import matplotlib.pyplot as plt

# Plotting the count of subjects using matplotlib
data['subject'].value_counts().plot(kind='bar')

# Adding labels and adjusting rotation
plt.xlabel("Subjects", fontsize=13)
plt.ylabel("Number of courses", fontsize=13)
plt.title("Number of courses per subject", fontsize=15)
plt.xticks(rotation=65)

# Display the plot
plt.show()
```

### Which courses has more lectures (free or paid)?

```python
lecture_counts = data.groupby('is_paid')['num_lectures'].sum()
print(lecture_counts)
```

```python
lecture_counts.plot(kind='bar', rot=0)
plt.title('Total number of Lectures for Free and Paid Courses')
plt.xlabel('Course Type')
plt.ylabel('Total Number of Courses')
plt.show()
```
