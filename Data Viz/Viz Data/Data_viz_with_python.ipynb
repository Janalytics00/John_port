## Importing the necessary library

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2 ,3]
y = [1, 4, 9]
plt.plot(x, y)
plt.show()
```

## Adding a title to the graph

```python
x = [1, 2 ,3]
y = [1, 4, 9]
plt.plot(x, y)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## Plotting multiple line

```python
x = [1, 2 ,3]
y = [1, 4, 9]
z= [10, 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## Adding a Legend statement to label our graph

```python
x = [1, 2 ,3]
y = [1, 4, 9]
z= [10, 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["this is y", "this is z"])
plt.show()
```

```python
# Analysis with the country data
data = pd.read_csv('countries.csv')
```

```python
data
```

```python
# Compare the population growth in the US and China
```

```python
# the code below gives us all the countries that are U.S, so we'll have a lot of false 
data.country == 'United States'
```

```python
# We're only selecting the countries that is U.S
us = data[data.country == 'United States']
us
```

```python
# We're only selecting the countries that is China assigning it to the variable, china
china = data[data.country == 'China']
china
```

```python
# Plotting year over population of U.S
plt.plot(us.year, us.population)
plt.show()
```

```python
# The 1e8 shows the number is still big, so let's divide the population by 1M i.e 10**6 to show result in million
plt.plot(us.year, us.population / 10**6)
plt.show()
```

```python
# Let's plot china's data ontop of this plot
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year china.population / 10**6)
plt.show()
```

```python
# adding legend and axis
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```
