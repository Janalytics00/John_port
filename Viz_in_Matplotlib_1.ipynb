```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ['A','B', 'C', 'D', 'E']
y = np.random.randint(1000, 9000, 5)
# Change the figure size
plt.subplots(figsize = (8,4))

# Create a bar chart
plt.bar(x,y)

# add labels and titles
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Category by Sales Using Column Chart')
plt.show()
```

```python
Categories = ['A','B', 'C', 'D', 'E']
Sales = np.random.randint(1000, 9000, 5)

# Plot the data
# plt.bar(Categories, Sales, color = ['red', 'blue', 'green', 'yellow', 'purple'])
# plt.bar(Categories, Sales, color = ['red', 'blue', 'green', 'yellow', 'purple'], width = 0.5)
# plt.bar(Categories, Sales, edgecolor = 'black', linewidth = 2)
plt.bar(Categories, Sales, alpha = 0.5) # 0 - 1
plt.xlabel('Categories')
plt.ylabel('Sales')
plt.title('Column chart Example')
plt.show()
```

```python
colors = ['red', 'blue', 'green', 'yellow', 'purple']
edgecolor = 'black'
linewidth = 1.5
plt.subplots(figsize = (8,4))
plt.bar(Categories, Sales, color = colors, edgecolor = edgecolor, linewidth = linewidth, alpha = 0.5)
plt.xlabel('Categories')
plt.ylabel('Sales')
plt.title('Column chart Example')

# Adding data labels to the columns
for i, v in enumerate(Sales):
  plt.text(i,v, str(v), color = 'red', ha = 'center', va = 'bottom')


plt.show()
```

```python
for i, v in enumerate(Sales):
  print(i, "and",v)
# What the code is saying, is "for" index, i and value, v each element in sales list, print, index and value
```

```python
data1 = [3,5,1,6,4]
data2 = [2,6,3,5,7]

labels = ['A', 'B', 'C', 'E', 'E']
plt.subplots(figsize = (8,4))
plt.bar(labels, data1, label = 'Data 1')
plt.bar(labels, data1, label = 'Data 2', bottom = data1)

plt.title("Column chart with legends and grid lines")

plt.xlabel('labels')
plt.ylabel('Data')

# adding a legend
plt.legend(loc = 9)

# adding grid lines to the chart
plt.grid(True)

plt.show()
```

```python
# Creating H- Barchart
fig, ax = plt.subplots(figsize = (6,4))
ax.barh(Categories, Sales, color = '#091B39')

# Addong data labels
for i, s in enumerate(Sales):
  ax.text(s, i, str(s), ha = 'left', va = 'bottom')

ax.set_xlabel('Category')
ax.set_ylabel('Sale')
ax.set_title('H - Bar Chart')
plt.show()
```

```python
# Creatiing column chart and horizontal bar chart together

# sample data
x = np.array(['Germany', 'India', 'USA', 'Canada', 'France'])
y1 = np.array([10,20,30,40,50])
y2 = np.array([15,25,35,45,55])

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (16, 4))

ax1.bar(x,y1, color = 'blue')
ax1.set_xlabel('Category')
ax1.set_ylabel('Values')
ax1.set_title('Column Chart')

ax2.barh(x,y2, color = 'orange')
ax2.set_xlabel('Category')
ax2.set_ylabel('Values')
ax2.set_title('Horizontal Bar Chart')

plt.tight_layout()

plt.show()
```

```python
# sample data
x = np.array(['Germany', 'India', 'USA', 'Canada'])
y1 = np.random.randint([1000,900,50, 5])
y2 = np.random.randint([1000,900,50, 5])
```

```python
# Define the x-axis positions for each bar
bar_width = 0.40
r1 = np.arange(len(x))
r2 = [x + bar_width for x in r1]

fig, ax = plt.subplots()

ax.bar(r1, y1, color = 'blue', width = bar_width, edgecolor = 'black', label = 'y1')
ax.bar(r2, y2, color = 'red', width = bar_width, edgecolor = 'black', label = 'y2')

ax.set_xlabel('Quarter')
ax.set_ylabel('Sales')
ax.set_title('Quarter Sales by Region')

ax.legend()

plt.show()
```

```python
labels = ['Category A', 'Category B', 'Category C', 'Category D']
values = [50,35,40,45]
errors = [5,2,3,4]

fig, ax = plt.subplots()

ax.bar(labels, values, yerr = errors, align = 'center', alpha = 0.5, ecolor = 'black', capsize = 10)

ax.set_xlabel('Category')
ax.set_ylabel('Values')
ax.set_title('Bar Chart with Error Bars')

plt.show()
```
